```python
# imports and data
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

data = pd.read_csv('data/classification_results/fine_tuned/acc.csv', index_col=0)
data = data.melt()
data['method'] = data['variable'].apply(lambda x: x.split(' ')[0])
data['signal_type'] = data['variable'].apply(lambda x: x.split(' ')[1])
data['band'] = data['variable'].apply(lambda x: x.split('(')[1].split(')')[0])
data = data.drop(columns=['variable'])
data = data.rename(columns={'value': 'accuracy'})
```

```python
# Using statsmodels formula API to fit the linear model
model = smf.ols(formula="accuracy ~ method * signal_type * band", data=data).fit()

# Displaying the summary of the model
model_summary = model.summary()
model_summary
```
